```python
import os
import wget
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
file = 'salaries.csv'
if os.path.isfile(file):
    print('file exists')
else:
    wget.download('https://raw.githubusercontent.com/codebasics/py/master/ML/9_decision_tree/salaries.csv', file)
```

```python
df = pd.read_csv(file)
```

```python
df.head()
```

```python
df.info()
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.preprocessing import LabelEncoder
```

```python
label_company = LabelEncoder()
label_job = LabelEncoder()
label_degree = LabelEncoder()
```

```python
df['company_id'] = label_company.fit_transform(df['company'])
df['job_id'] = label_job.fit_transform(df['job'])
df['degree_id'] = label_degree.fit_transform(df['degree'])
```

```python
df.head()
```

```python
X = df[['company_id', 'job_id', 'degree_id']]
y = df['salary_more_then_100k']
```

```python
tree_clf = DecisionTreeClassifier()
```

```python
tree_clf.fit(X, y)
```

```python
tree_clf.score(X,y)
```

```python
from sklearn.tree import plot_tree
```

```python
fig = plt.figure(figsize=(20,10))
_ = plot_tree(tree_clf, feature_names=['company','job','degree'], class_names=['gt 100k','lt 100k'], filled=True)
```

```python
file = 'titanic.csv'
if os.path.isfile(file):
    print('file exists')
else:
    wget.download('https://raw.githubusercontent.com/codebasics/py/master/ML/9_decision_tree/Exercise/titanic.csv', file)
```

```python
df = pd.read_csv(file)
```

```python
df.head()
```

```python
df.info()
```

```python
df.columns = df.columns.str.lower()
```

```python
df.drop(columns=['passengerid','name','ticket'], inplace=True)
```

```python
df.drop(columns=['cabin', 'parch'], inplace=True)
```

```python
df.drop(columns=['sibsp','embarked'], inplace=True)
```

```python
df['sex'].value_counts()
```

```python
pd.crosstab(df['survived'], df['sex'])
```

```python
df[df['sex'] == 'female']['age'].median()
```

```python
df[df['sex'] == 'male']['age'].median()
```

```python
df['age'].median()
```

```python
df['age'].fillna(df['age'].median(), inplace=True)
```

```python
from sklearn.preprocessing import LabelEncoder
label = LabelEncoder()
```

```python
df['sex'] = label.fit_transform(df['sex'])
```

```python
df.head()
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn import metrics
```

```python
X = df.drop(columns=['survived'])
y = df['survived']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=50)
```

```python
tree_clf = DecisionTreeClassifier()
```

```python
tree_clf.fit(X_train, y_train)
```

```python
y_pred = pd.Series(tree_clf.predict(X_test))
```

```python
tree_clf.score(X_test, y_test)
```

```python
sns.heatmap(metrics.confusion_matrix(y_test, y_pred), annot=True)
plt.xlabel('Predcited')
plt.ylabel('True')
```

```python
metrics.balanced_accuracy_score(y_test, y_pred)
```

```python
fig = plt.figure(figsize=(20,15))
_ = plot_tree(tree_clf, filled = True)
```
